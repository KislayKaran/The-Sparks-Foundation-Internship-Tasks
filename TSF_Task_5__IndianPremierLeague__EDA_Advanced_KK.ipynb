# **Total no of Fours and Sixes hit by each Teams**

```python
# we will print numbers of fours hit by team
four_data=df[df['batsman_runs']==4]
four_datas=four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])

# we will print no of sixes hit by team
six_data=df[df['batsman_runs']==6]
six_datas=six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])

df_fs= pd.merge(four_datas,six_datas, left_on='batting_team', right_on ='batting_team')
df_fs
```

```python
# we will plot graph on fours and sixes hit by players
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
batsman_fs=pd.merge(batsman_four,batsman_six, left_on='batsman', right_on ='batsman')
batsman_fs.sort_values('six',ascending=0).head(20)
```

# **Top 10 Batsmen with the best strike rates over the years**

```python
no_of_balls = pd.DataFrame(df.groupby('batsman')['ball'].count()) #total number of matches played by each batsman
runs = pd.DataFrame(df.groupby('batsman')['batsman_runs'].sum()) #total runs of each batsman
seasons = pd.DataFrame(df.groupby('batsman')['season'].nunique()) #season = 1 implies played only 1 season

batsman_strike_rate = pd.DataFrame({'balls':no_of_balls['ball'],'run':runs['batsman_runs'],'season':seasons['season']})
batsman_strike_rate.reset_index(inplace = True)

batsman_strike_rate['strike_rate'] = batsman_strike_rate['run']/batsman_strike_rate['balls']*100
highest_strike_rate = batsman_strike_rate[batsman_strike_rate.season.isin([10,11])][['season','batsman','strike_rate']].sort_values(by = 'strike_rate',
                                                                                                            ascending = False)

highest_strike_rate.head(10)
```

```python
plt.figure(figsize = (20,6))
sns.barplot(x='batsman', y='strike_rate', data = highest_strike_rate.head(10), hue = 'season')
plt.title("Highest strike rates in IPL",fontsize= 30, fontweight = 'bold')
plt.xlabel("Player", size = 25)
plt.ylabel("Strike Rate", size = 25)
plt.xticks(size = 14)
plt.yticks(size = 14)
```

# **Which bowlers have picked up wickets more frequently?**

```python
#strike_rate = balls bowled by wickets taken
balls_bowled = pd.DataFrame(df.groupby('bowler')['ball'].count())
wickets_taken = pd.DataFrame(df[df['dismissal_kind'] != 'no dismissal'].groupby('bowler')['dismissal_kind'].count())
seasons_played = pd.DataFrame(df.groupby('bowler')['season'].nunique())
bowling_strike_rate = pd.DataFrame({'balls':balls_bowled['ball'],'wickets':wickets_taken['dismissal_kind'],
                          'season':seasons_played['season']})
bowling_strike_rate.reset_index(inplace = True)
```

```python
bowling_strike_rate['bowling_strike_rate'] = bowling_strike_rate['balls']/bowling_strike_rate['wickets']
def highlight_cols(s):
    color = 'skyblue'
    return 'background-color: %s' % color
#Strike rate for bowlers who have taken more than 50 wickets
best_bowling_strike_rate = bowling_strike_rate[bowling_strike_rate['wickets'] > 50].sort_values(by = 'bowling_strike_rate', ascending = True)
best_bowling_strike_rate.head().style.applymap(highlight_cols, subset=pd.IndexSlice[:, ['bowler', 'wickets','bowling_strike_rate']])
```

# **Bowlers with best Economy Rates**

```python
overs_bowled = pd.DataFrame(df.groupby('bowler')['over'].count())
runs_conceded = pd.DataFrame(df.groupby('bowler')['total_runs'].sum())
bowling_eco_rate = pd.DataFrame({'overs_bowled':overs_bowled['over'],'runs_conceded':runs_conceded['total_runs']})
bowling_eco_rate.reset_index(inplace = True)
```

```python
bowling_eco_rate['economy_rate'] = bowling_eco_rate['runs_conceded']/bowling_eco_rate['overs_bowled']
best_bowling_eco_rate = bowling_eco_rate[bowling_eco_rate['overs_bowled']> 100 ].sort_values(by = 'economy_rate', ascending = True)
best_bowling_eco_rate.head(20).style.applymap(highlight_cols, subset=pd.IndexSlice[:, ['bowler','overs_bowled', 'runs_conceded','economy_rate']])
```
